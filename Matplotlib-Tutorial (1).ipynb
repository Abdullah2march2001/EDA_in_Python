# Customising Graphs

```python
#Adding multiple lines in one graph

plt.plot([2,3,4,5],[1,3,4,6])
plt.plot([1,2,3,5],[1,3,4,6])

plt.show()
```

```python
#Adding data with numpy
x = np.arange(1,5)

plt.plot(x,x)
plt.plot(x,x+3)
plt.plot(x,x+9)

plt.show()
```

Legend, Grid and Axes
LEGEND: for location of legend available options are: 'best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center'

AXIS: Axis command takes 4 values [xstart,xfinal,ystart,yfinal]

Line and Marker Customisation
Available Shapes for Markers are:

"s"=square, "p"=pentagon, "."=point, "o"=circle, "d"=diamond, etc.

Available Lines are:

"-", ":", "--", "-.", etc.


By Default, third argument in plot() function is line pattern alongside other customizations like `color, markersize, linestyle, linewidth, marker, markeredgecolor, etc`
More can be read at:
LineStyles: https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html

MarkerStyle: https://matplotlib.org/stable/gallery/lines_bars_and_markers/marker_reference.html

```python
#Customize the graph 

plt.plot(x,x, color='purple',linestyle =':', linewidth =3.0, marker ='d', markeredgecolor="green")
plt.plot(x,x+3, color='orange',linestyle =':', linewidth =3.0, marker ='d', markeredgecolor="green")
plt.plot(x,x+9, color='red',linestyle =':', linewidth =3.0, marker ='d', markeredgecolor="green")

#title of graph
plt.title("Line Plot")

#Labels
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

#legends
plt.legend(["1st","2nd","3rd"], loc="best")

#grid
plt.grid("true")

# Set the axis limits
plt.axis([0, 5, 1, 15])


plt.show()
```

# Subplots to create multiple plots

```python
# Create a figure
fig = plt.figure(figsize=(30, 15), facecolor='cyan')

# Subplot syntax => fig.add_subplot(rows, columns, position)
rows = 2
columns = 3

# Follows a grid format => rows x columns
# Create charts on the figure using add_subplot
chart1 = fig.add_subplot(rows, columns, 1)
chart2 = fig.add_subplot(rows, columns, 2)
chart3 = fig.add_subplot(rows, columns, 3)

# Each of the above subplots acts as a separate plot and can be used to plot anything in space
x = np.arange(1, 5)
chart1.plot(x, x, ":o",color="red")
chart2.plot(x, x + 2, "-s")
chart3.plot(x, x + 3, "-.p")


# Plot on subplots
plt.show()
```

# Pie-Chart and Exploded Pie-Chart

```python
# Sample data
labels = ['Peshawar', 'Karachi', 'Islamabad', 'Lahore']
sizes = [15, 30, 45, 10]

# Plot pie chart with values and labels
plt.pie(sizes, labels=labels, autopct="%1.1f%%")
plt.legend(labels,loc="upper right")
plt.title("Pie Chart")

plt.show()
```

```python
# Sample data
labels = ['Peshawar', 'Karachi', 'Islamabad', 'Lahore']
sizes = [15, 30, 45, 10]

exp = [0.1,0.02,0.05,0.1]
# Plot pie chart with values and labels
plt.pie(sizes, labels=labels, autopct="%1.1f%%",explode=exp)
plt.legend(labels,loc="upper right")
plt.title("Pie Chart")

plt.show()
```

# Bar-Charts

```python
plt.figure(figsize=(7,5), facecolor='cyan')

# labels and values for bar chart
labels = ['India', 'USA', 'Britain', 'Germany', 'Japan']
values = [25, 30, 20, 10, 35]


# Plot the bar chart
plt.bar(labels,values , color = "orange" ,width=0.5)

plt.title("Single Bar Chart")
plt.xlabel("Countries")
plt.ylabel("Values")

plt.show()
```

# Grouped Bar Charts

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(7, 5), facecolor='cyan')
plt.title("Group Bar Chart")
# Labels and values for grouped chart
labels = ["India", "USA", "Britain", "Germany", "Japan"]
values = [25, 30, 20, 10, 35]
values2 = [30, 20, 15, 15, 40]

# Specify Label Positions for tickers on the x-axis
positions = np.arange(len(labels))

# Draw 2 bar charts for displaying grouped bars alongside widths
plt.bar(positions, values, color="red", width=0.3)
plt.bar(positions + 0.3, values2, color="blue", width=0.3)

# Give position to display x tickers on the X-axis, which should be the position of the bar + half of the width of bars
plt.xticks(positions + 0.15, labels)

plt.show()
```

# Horizontal Bar Chart 

```python
# Labels and values for the horizontal bar chart
labels = ["India", "USA", "Britain", "Germany", "Japan"]
values = [25, 30, 20, 10, 35]
values2 = [30, 20, 15, 15, 40]

# Create positions for the bars
positions = range(len(labels))

# Set the figure size
plt.figure(figsize=(7, 5), facecolor='cyan')

# Create horizontal bar charts
plt.barh(positions, values, color="red", height=0.3, label='Values 1')
plt.barh([pos + 0.3 for pos in positions], values2, color="blue", height=0.3, label='Values 2')

# Set the y-axis labels
plt.yticks([pos + 0.15 for pos in positions], labels)

# Set labels and a legend
plt.xlabel('Values')
plt.title('Horizontal Bar Chart')
plt.legend()

# Show the chart
plt.show()
```

# 3D Scatter Plots

```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]
z = [10, 20, 30, 40, 50]

# Define colors and size for each point
colors = ['r', 'g', 'b', 'c', 'm']
size = [30, 40, 50, 60, 70]

# Create a figure and a 3D subplot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Use the scatter function to plot the 3D data
ax.scatter(x, y, z, c=colors, s=size, marker='^')

# Set labels for the axes
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

# Display the 3D scatter plot
plt.show()
```

# 3D Bar Plots

```python
# create a 3d projection subplot to display our scatter plot
fig = plt.figure()
ax = fig.add_subplot(1,1,1, projection = '3d')
# define distances from plane
x =[1,2,3,4,5]
y =[3,6,2,5,7]
# changing this would result in bar uplifting from plane
z =[0,0,0,0,0]

# length of bar in direction plane
dx =[1,1,1,1,1]
dy =[1,1,1,2,1]
dz =[1,2,6,3,3]

# plot the 3d bar graph using both distances and lengths of bar
ax.bar3d(x,y,z, dx,dy,dz, color='gold' )

plt.show()
```
